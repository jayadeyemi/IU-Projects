### **Problem description:**
        In the fictional setting of Bitland, a recent construction mishap has resulted in the diversion of a left-side road to the right, creating a challenging scenario on the highway. Cars on this highway, moving northward or southward at uniform speed, cannot turn back. To effectively manage this situation, the objective is to rescue as many people as possible who are traveling in these cars. Each car is represented by an integer where the sign indicates its direction (positive for north, negative for south) and the magnitude represents the number of people in the car.

        The task is to create a function, saveMaximumPeople(cars), that determines the maximum number of people that can be saved from potential collisions. When cars moving in opposite directions meet, they will inevitably collide. In such cases, to save the maximum number of people, the car with fewer occupants should be considered "sacrificed", and its occupants will not be counted in the final tally.


Constraints:

        • Solve the question using stack
        • Expected time complexity is O(n)

Example:
<div style="text-align: center;">
<img src="Images/Savemaximumpeople.jpg" alt="Illustration of Example 1" />
</div>
        cars = [4,-2,10,6,9,-4]
        saveMaximumPeople(cars) = 29


Explanation: 

        The car at index 1 (carrying 2 people south) will collide with the car at index 0 (carrying 4 people north). Choosing to save the occupants of the car at index 0 is optimal. Likewise, the car at index 4 will collide with the car at index 5, and choosing to save the occupants of car at index 4 maximizes the number saved. No collision affects the cars at indexes 2 and 3, and they continue safely.

Example:

        cars = [1,1,1,1,1,1,1,1,1]
        saveMaximumPeople(cars) = 9

_Explanation:_

        As all cars are moving in the same direction, there are no collisions, and all occupants are saved.



```python
def saveMaximumPeople(cars):
    saved = []
    previous = cars.pop()
    saved.append(previous)
    while len(cars) > 0:
        current = cars.pop()
        
        if current == abs(previous):
            if saved:
                saved.pop()
            continue
        elif current > 0:
            if previous < 0:
                if saved:
                    saved.pop()
                saved.append(max(current, abs(previous)))
            else:
                saved.append(current)
        elif current < 0:
                saved.append(abs(current))
        
        print(saved)
        previous = current
    sum_saved = sum(saved)
    return sum_saved
```

```python

def ironman_vs_thanos_again(s):
    stack = []  # Initialize an empty stack to store 'A's
    back_stack = []  # Initialize an empty stack
    longest_balanced = 0  # Initialize the length of the longest balanced part
    max_balanced = 0  # Initialize the length of the maximum balanced part  
    last_item = None
    for char in s:
        if char == 'A':
            stack.append(char)  # Push 'A' onto the stack
            back_stack.append(char)
        elif char == 'T' and stack:  # If 'T' is encountered and stack is not empty
            back_stack.append(char)
            stack.pop()  # Pop matching 'A' from the stack
            longest_balanced += 2  # Increment the length of the longest balanced part
        else:
            longest_balanced = 0

        max_balanced = max(longest_balanced, max_balanced)
        print(stack, longest_balanced, max_balanced) 
    if stack and back_stack[-1]=='T':
        max_balanced -= 2
    return max_balanced
```

```python

def impatientBob(Prices: list[int]) -> list[int]:
    n = len(Prices)
    stack = []
    result = [0] * n
    
    for i in range(n):
        while stack and Prices[i] > Prices[stack[-1]]:
            idx = stack.pop()
            result[idx] = i - idx
        stack.append(i)
    
    return result
```

```python
def scoreOfMagicalString(s: str) -> int:
    stack = []
    cur_score = 0
    
    for char in s:
        if char == '(':
            stack.append(cur_score)
            cur_score = 0
        else:  # char == ')'
            cur_score = stack.pop() + max(cur_score * 2, 1)
    
    return cur_score
```
