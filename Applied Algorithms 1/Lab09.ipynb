```
def IdYaKnYiK(people:list[int])->int:
    # use merge sort to sort the people at O(nlogn)
    def mergeSort(people:list[int])->list[int]:
        if len(people) <= 1:
            return people
        mid = len(people) // 2
        left = mergeSort(people[:mid])
        right = mergeSort(people[mid:])
        return merge(left, right)
    def merge(left:list[int], right:list[int])->list[int]:
        result = []
        i = j = 0
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
        result += left[i:]
        result += right[j:]
        return result
    people = mergeSort(people)

    # minimum cost to balance the people at O(n)
    minCost = 0
    minlist = []

    n = len(people)
    for i in range(n // 2):
        # since the people are sorted, the minimum cost is the sum of the difference between the first and last element
        minCost += (people[n - i - 1] - people[i])


    return minCost

def heapify(arr, n, i):
    print("Heapifying with root at index:", i)
    largest = i  
    l = 2 * i + 1     
    r = 2 * i + 2     
    
    print("Left child index:", l)
    print("Right child index:", r)

    if l < n and arr[l] > arr[largest]:
        largest = l

    if r < n and arr[r] > arr[largest]:
        largest = r

    if largest != i:
        print("Swapping", arr[i], "with", arr[largest])
        arr[i], arr[largest] = arr[largest], arr[i]  
        print("After swapping:", arr)
        heapify(arr, n, largest)
    else:
        print("No swap needed. Heap property already satisfied.")

# Example usage:
arr = [12, 11, 13, 5, 6, 7]
n = len(arr)

print("Original array:", arr)
for i in range(n // 2 - 1, -1, -1):
    heapify(arr, n, i)

print("Array after heapify:", arr)
```
